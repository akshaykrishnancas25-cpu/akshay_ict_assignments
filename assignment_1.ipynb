1. Grocery Billing
A shopkeeper wants to calculate the total bill for two items. Write a program that takes
the price and quantity of each item and prints the total bill amount.
Sample Input:
Item1 price: 45
Item1 quantity: 3
Item2 price: 30
Item2 quantity: 2
Sample Output:
Total Bill: 195

```python
item1=int(input('Enter the price of item1 '))
item2=int(input('Enter the price of item2 '))
q1=int(input('Enter quantity of item 1 '))
q2=int(input('Enter quantity of item2 '))
total=(item1*q1+item2*q2)
print('total price is=',(total))
```

3. Salary Slip
Given the basic salary of an employee, calculate the total salary including HRA (20%) and
DA (10%).
Sample Input:
Enter basic salary: 20000
Sample Output:
Net Salary: 26000.0

```python
s = int(input('enter your salary '))
hra = int(input('enter your hra' ))
da = int(input('enter your da '))
HRA=s*hra/100
DA=s*da/100
total=s+HRA+DA
print('net salary=',(total))
```

8. Area of a Triangle (Using Heron‚Äôs Formula)
A land surveyor wants to calculate the area of an irregular triangular plot.
Write a Python program to find the area of a triangle when all three sides are known,
using Heron‚Äôs formula:
ùë† =
ùëé + ùëè + ùëê
2
Area = ‚àöùë†(ùë† ‚àí ùëé)(ùë† ‚àí ùëè)(ùë† ‚àí ùëê)
Sample Input:
Enter side a: 5
Enter side b: 6
Enter side c: 7
Sample Output:
Area of the triangle: 14.70




```python
p=int(input('enter the first side ' ))
q=int(input('enter the second side ' ))
r=int(input('enter the third side ' ))
s=(p+q+r)/2
area=(s*(s-p)*(s-q)*(s-r))**0.5
print('area of triangle',(area))
```

9. Paint Cost Estimator
A painter charges ‚Çπ25 per square meter. Write a program to calculate the total cost to
paint a rectangular wall. The wall has length and height, and there are two windows each
measuring 1.5m √ó 1.2m that should not be painted. Find the total area to be painted and
its total cost.
Sample Input:
Enter wall length (in meters): 12
Enter wall height (in meters): 8
Sample Output:
Total area to be painted: 92.4 sq.m
Total painting cost: ‚Çπ2310.0

```python
p=int(input('enter the length of wall '))
q=int(input('enter the height of wall '))
area=(p*q)-1.5*1.2*1.5*1.2
cost=area*25
print('total area to be painted',(area))
print('total painting cost',(cost))
```

10. Electricity Bill Calculator
An electricity board charges consumers based on unit consumption. The rate per unit is
‚Çπ5.2. Additionally, there is a fixed charge of ‚Çπ100 on every bill. Write a Python program
that takes the number of units consumed and calculates the total bill amount.
Sample Input:
Enter number of units consumed: 250
Sample Output:
Electricity Bill: ‚Çπ1400.0

```python
p=int(input('enter the number of units consumed '))
rate=int(input('enter the rate per unit '))
fixed=int(input('enter the fixed charge '))
total=p*rate+fixed
print('electricity bill',(total))
```
